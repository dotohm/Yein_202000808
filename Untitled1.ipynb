```python
import nltk
nltk.download('brown', 'inaugural')
```

```python
brown = nltk.corpus.brown
```

```python
romance = brown.raw(categories = 'romance')[:5000]
romance
```

```python
sent = romance.split('.')
sent
```

```python
sents = [w.split(' ') for w in sent]
sents
```

```python
for i in sents:
    [w.split('/') for w in i]
    print(i)
```
