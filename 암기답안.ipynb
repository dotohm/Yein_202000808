# 문제 2

```python
#1.
f = open("HarryPotter.txt", "r", encoding='utf-8')
data = f.read()
f.close()

# 2.소제목을 분류하기 위해 replace 16개 추가함
data = data.replace('Mr.', 'Mr').replace('Mrs.', 'Mrs').replace('J.K.', 'J*K*').replace('?', '.').replace('!','.').replace('.”','.').replace('LIVED', 'LIVED.').replace('SS','SS.').replace('NO ONE', 'NO ONE.').replace('YS', 'YS.').replace('EY', 'EY.').replace('RS','RS.').replace(' HAT',' HAT.').replace(' MASTER',' MASTER.').replace('UEL','UEL.').replace('EEN','EEN.').replace('DITCH', 'DITCH.').replace('SED', 'SED.').replace('BACK','BACK.').replace('REST','REST.').replace('OOR','OOR.').replace('CES', 'CES.')
sentences = [sent.strip() for sent in data.split('.')]
sentences_2 = [w.replace('J*K*', 'J.K.').replace('\n\n','') for w in sentences]

#3.
nospace_words = [word for word in sentences if ' ' not in word]
sent = [w + ' ' for w in nospace_words if '' != w]

import string
punct = string.punctuation

sent_tokens = [[w for w in sentence.split() \
                   if w not in punct and w]
                      for sentence in sentences_2]

#4.
output = []
for sentence in sent_tokens:
    temp = []
    for word in sentence:
        # 문장부호 아닌지 다시 확인
        if word not in punct and word: # not empty
            if "’" in word:
                # She's -> [She, 's] 
                # fur-pieces's -> [fur-pieces, 's]
                index = word.find("’")
                
                if word[index-2:] == 'dn’t':
                    word = word.replace('dn’t', 'd not')
                elif word[index-2:] == 'an’t':
                    word = word.replace('an’t', 'an not')
                elif word[index-3:] == 'won’t':
                    word = word.replace('won’t', 'will not')
                elif word[index-1:] == 'n’t':
                    word = word.replace('n’t', ' not')
                elif word[index:] == '’m':
                    word = word.replace('’m', 'am')
                elif word[index:] == '’ve':
                    word = word.replace('’ve', 'have')
                elif word[index:] == '’ll':
                    word = word.replace('’ll', 'will')
                elif word[index:] == '’re':
                    word = word.replace('’re', 'are')
                words = word[:index], word[index:]
            else:
                words = [word]
        
            for word in words:
                if '-' in word:
                    index = word.find('-')
                    words = word[:index], word[index+1:]
                    if '-' in word[index+1:]:
                        index_2 = word[index+1:].find('-')
                        words = word[:index], word[index+1:index + index_2+1], word[index+index_2+2:]
                        if '-' in word[index+1+index_2+1:]:
                            index_3 = word[index+1+index_2+1:].find('-')
                            words = word[:index], word[index+1:index+index_2+1], word[index+index_2+2:index+index_2+index_3+1], word[index+index_2+index_3+2:]
                else:
                    words = [word]     
            temp.extend(words)       
    output.append(temp)
    
output_list = []

import string
string = string.punctuation + '”' + '“' + '—' + "’" + '...' + '•' + '\n' + '\t' +'■'
for w in output:
    for r in string:
        w = [x.replace(r, '') for x in w]
    output_list.append([x for x in w if w]) 

import nltk
stopwords = nltk.corpus.stopwords.words('english')

#5.
result = []
for word in output_list:
    for w in word:
        if w not in stopwords:
            w = w.lower()
            result.append(w)

fdist = nltk.FreqDist(result)
fdist.plot(50, cumulative=True)

#6.
total = len(result)
total
```

```python
def print_gugu(dan):
    print(f'=== {dan}단 ===')
    for i in range(1, 10):
        print(f'{dan} X {i} = {dan * i:2d}')
    print()
    
def check_dan(n, m = '9'):
    if not(1 <= int(n) <= 9) or not(1 <= int(m) <= 9):
        return False
    return True

def end_program():
    print('이용해주셔서 감사합니다.')
    
def gugudan():

    print("""
    구구단 출력기
    ----------------------------------------------------------------------
            1) n단 출력    2) n단까지 출력   3) n ~ m단 출력   q) 나가기
    ----------------------------------------------------------------------
    """)
    menu = input('메뉴를 선택하세요 > ')

    if menu == '1':
        print('1) n단 출력')
        dan_input = True
        while dan_input:
            dan = input('>>> 몇 단을 출력할까요? ')
            if dan == 'q':
                return end_program()
            
            elif dan.isnumeric() and check_dan(dan):
                print_gugu(int(dan))
                dan_input = False
            else:
                print('구구단은 1~9단 까지만 출력할 수 있습니다. 다시 입력하세요.')
        return gugudan()
    
    elif menu == '2':
        print('2) n단까지 출력')
        dan_input = True
        while dan_input:
            dan = input('>>> 구구단을 몇 단까지 출력할까요? ')
            if dan == 'q':
                return end_program()
            
            elif dan.isnumeric() and check_dan(dan):
                for i in range(1, int(dan) + 1):
                    print_gugu(i)
                dan_input = False
            else:
                print('구구단은 1~9단 까지만 출력할 수 있습니다. 다시 입력하세요.')
        return gugudan()
    
    elif menu == '3':
        print('3) n ~ m단 출력')
        dan_input = True
        while dan_input:
            n = input('>>> 몇 단부터 출력할까요? ')
            m = input('>>> 몇 단까지 출력할까요? ')
            if (n == 'q') or (m == 'q'):
                return end_program()
            
            elif n.isnumeric() and m.isnumeric() and check_dan(n, m):
                if m < n:
                    print(f'시작하는 단이 끝나는 단보다 작거나 같아야 합니다. 다시 입력하세요.')
                else:
                    for i in range(int(n), int(m) + 1):
                        print_gugu(i)
                    dan_input = False
            else:
                print('구구단은 1~9단 까지만 출력할 수 있습니다. 다시 입력하세요.')
        return gugudan()
    
    elif menu != 'q':
        print('잘못 입력하셨습니다. 초기 화면으로 돌아갑니다.')
        return gugudan()
    
    if menu == 'q':
        return end_program()
    
gugudan()
```

```python
def text_normalize(words_token):
    import string
    
    APOSTROPHE, HYPHEN = '\'', '-'
    
    # 기존에 없는 문장부호 추가
    punct = list(string.punctuation)
    my_punct = set([word[0] for word in words_token
                    if (not word[0].isalnum()) and (word not in punct)])
    punct.extend(list(my_punct))
    # 1) 문장부호, 공백제거
    words_token = [word.strip() for word in words_token if word[0].isalnum()]
    
    data = []
    for word in words_token:
        if word not in punct:
            # 2) apostrophe 처리 (불용어 처리 용이를 위해 split)
            if APOSTROPHE in word:
                words = word.split(APOSTROPHE)
            else: 
                words = [word]
            # 3) hyphen 처리 (불용어 처리 용이를 위해 split)
            if HYPHEN in word:
                words = word.split(HYPHEN)
            
            # 처리한 words를 data에 추가하기
            data.extend(words)
            
    # 4) 불용어와 빈 단어 '' 등 기타 불필요한 토큰 제거
    stop_words = nltk.corpus.stopwords.words('english')
    output = [word for word in data 
              if (not word.lower() in stop_words) and word.isalnum()]
    
    return output
```
