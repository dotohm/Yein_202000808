#### "편의점에서 고객이 물건을 산다"

 

- 클래스를 활용하여 위의 목표를 달성한다.
    - 구성할 클래스: 사람, 편의점, 물건
- 포함할 주요 내용:
    - 편의점에 물건 등록한다.
    - 일정 물건에 할인 이벤트를 만든다.
    - 수입을 계산한다.
    - 고객 명단을 관리한다.
    
    - 고객이 물건을 구매한다.
    - 고객의 포인트가 적립된다.
    - 일을 해서 돈을 번다.(고객의 수입을 말하는 것 같고)
    - 잔액을 조회한다.(그 수입에서 오늘 지불한 값을 뺀 거를 구하라는 것 같아)
- 반드시 활용해야 할 개념
    - 캡슐화
    - 추상클래스
    - 클래스메서드
    - 상속
    - 네임드튜플 혹은 데이터클래스
- 중요한 것: 최대한 상상력을 발휘하여 상호 관계를 구현할수록 점수 높을 것!

```python
class Product:
    def __init__(self, product_code, name, price):
        self.product_code = product_code
        self.name = name
        self.price = price
        
class Store:
    def __init__(self):
        self.product_list = []
        self.cart = []
    
    def add(self, product):
        self.product_list.append(product)
        
    def display(self):
        for product in self.product_list:
            print(f'{product.product_code}\t {product.name} \t{product.price}')
            
    def add_to_cart(self, product):
        for item in self.product_list:
            if item.product_code == product:
                self.cart.append(item)
                print(f'장바구니에 {item.name}이 추가되었습니다')
                return self.cart
        print('더이상 물건이 없습니다')
        
    def remove_from_cart(self, product):
        for item in self.cart:
            if item.product_code == product:
                self.cart.remove(item)
                print(f'{item.name}이 카트에서 삭제됐습니다.')
                return
        print('해당 번호는 카트에 없습니다.')
        
    def generate_bill(self):
        amount = 0
        for product in self.cart:
            amount += product.price
        print(f'총 {amount}원 입니다.')
        
    def display_cart(self):
        for item in self.cart:
            print(f'{item.name}')
while True:       
    s = Store()
    s.add(Product(1, '우유', 20))
    s.add(Product(2, '계란', 30))
    s.add(Product(3, '식빵', 10))
    s.add(Product(4, '쌀', 200))
    print('1. 상품 보기\n2. 장바구니에 담기 \n3. 상품 삭제하기\n4. 장바구니 보기\n5. 계산하기\n6. 나가기')
    choice = int(input('번호를 입력하세요'))
    if choice == 1:
        s.display()
    elif choice == 2:
        code = int(input('상품 번호를 입력하세요: '))
        s.add_to_cart(code)
    elif choice == 3:
        code = int(input('삭제할 상품 번호를 입력하세요: '))
        s.remove_from_cart(code)
    elif choice == 4:
        s.display_cart()
    elif choice == 5:
        s.generate_bill()
    elif choice == 6:
        print('감사합니다')
        break
    else:
        print('보기에 있는 번호를 눌러주세요')
```
